## 2.Dictionary Reprsentaion
as in sparse representaion we was wasting alot of memory only for storing zeros which are not adding any useful infromation so we dont have to store it all node will be keys in dictionary and value will be list of all node to which it is connected  so we will get rid of all zeros and this will be extremely efficient in terms of storage and retreival

```python
class DiGraph:
    def __init__(self):
        self.g={}
    def add_node(self,node):
        if node not in self.g:
            self.g[node]=[]
    def add_edge(self,src,dest):
        if src in self.g and dest in self.g and dest not in self.g[src]:
            self.g[src].append[dest]
        elif src not in self.g:
            raise KeyError(str(src)+" not in graph")
        elif src not in self.g:
            raise KeyError(str(dest)+" not in graph")
```
