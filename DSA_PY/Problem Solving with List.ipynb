```python
def min_lst(lst):
    if len(lst)==0:
        return None
    min1=lst[0]
    idx=0
    for i in range(1,len(lst)):
        if lst[i]<min1:
            min1=lst[i]
    return min1
```

```python
def three_highest(lst):
    if len(lst)==0:
        return None
    if len(lst)<=3:
        return lst
        
    high1,high2,high3=lst[0],lst[0],lst[0]
    for i in lst[1:]:
        if i >=high1:
            high3=high2
            high2=high1
            high1=i
        elif i >= high2:
            high3=high2
            high2=i
        elif i>high3:
            high3=i
    return (high1,high2,high3)
```
