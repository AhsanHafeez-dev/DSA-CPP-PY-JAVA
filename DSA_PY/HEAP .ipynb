### Mathematical equation for resolving childs
if node is at nth index<br>
its left  child will be on <b><i>2n+1</i></b> index<br>
its right child will be on <b><i>2n+2</i></b> index<br>

```python
def heapify(lst):
    ln=len(lst)
    for i in range(ln):
        l=(2*i)+1
        r=(2*i)+2
        if l<ln and r<ln:
            m=max(lst[l],lst[i],lst[r])
            if lst[i]==m:
                pass
            elif lst[r]==m:
                lst[r],lst[i]=lst[i],lst[r]
            else:
                lst[l],lst[i]=lst[i],lst[l]
        if l<ln:
            lst[l],lst[i]=min(lst[l],lst[i]),max(lst[l],lst[i])
        if r<ln:
            lst[r],lst[i]=min(lst[r],lst[i]),max(lst[r],lst[i])
        
```

```python
def swap(lst,i,j):
    lst[i],lst[j]=lst[j],lst[i]
```

```python
def heapify(lst,root,n,fn):
    desired=root
    l=(2*root)+1
    r=(2*root)+2
    if l<n and fn(lst[l],lst[desired]):
        desired=l
    if r<n and fn(lst[r],lst[desired]):
        desired=r
    if desired!=root:
        swap(lst,desired,root)
        heapify(lst,desired,n,fn)
```

```python
def comparison_max(a,b):
    return a>b
def comparison_min(a,b):
    return a<b
```

```python
def build_heap(lst,fn):
    n=len(lst)
    for i in reversed(range(n//2)):
        heapify(lst,i,n,fn)
```

```python
def heapify(n,lst,root,cmp):
    left  = 2 * root + 1
    right = 2 * root + 2
    largest=root
    
    if left<n and cmp(lst[left],lst[largest]):
        largest = left
    
    if right<n and cmp(lst[right],lst[largest]):
        largest = right
        
    if largest != root:
        lst[largest],lst[root] = lst[root],lst[largest]
        heapify(n,lst,root,cmp)
        
    
        
        
```
