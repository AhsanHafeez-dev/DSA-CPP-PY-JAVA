# STACK 
have you seen books pile up on one another ? if you want to retrieve or pick some book which one will you get?
ofcourse no matter which book you want you have to remove top than lower book all till your desired location
suppose book are in cabinet in which you can see or grab mid books
so how you are going to implement this logic in code
as list we have seen till now follow no proper sequence or order to insert or retrieve data
how are you gonna restrict this sequence
solution is <mark> STACK </mark>
stack is LIFO(last in first out ) data structure
ELEMENT WHICH IS INSERTED LAST WILL BE RETRIEVE FIRST
you cannot view any element rather than lastone in stack (by defination)


```python
class Stack:
    
    def __init__(self):
        
        self.stack=[]
        
    def push(self,val):
        
        self.stack.append(val)
        
    def pop(self):
        
        if len(self.stack)==0:
            print("StackOverflow !")
            return
        
        return self.stack.pop()
    
    def peek(self):
        
        return self.stack[-1]
    
    def is_empty(self):
        
        return len(self.stack)==0
```

```python
s=Stack()

l=10
x=""
p=""

for i in range(10):
    s.push(i)
    x+= str(i)+','
    
x=x.rstrip(',')

for i in range(l):
    p+=str(s.pop())+","
    
p=p.rstrip(',')    

print("push order : ",x)
print("pop  order : ",p)
```

# APPLICATION OF STACK
## Bracket Matching
whenever you write expression in any language it is being checked for different syntactical thing before evaluation
one of these checks is bracket matching when you see wether all types of brackets used in expression are in proper order 
and are closed properly.it is generally implmented using stack


```python
def bracket_matching(expression):
    
    dic={
        "(":")",
        "{":"}",
        "[":"]",        
    }
    
    s=Stack()
    
    for char in expression:
        if char in dic.keys():
            s.push(dic[char])
        
        if char in dic.values():
            
            if s.is_empty() or char != s.peek():
                return False
                
            else:
                s.pop()
                
    return s.is_empty()
            
```

```python
exp="{([])}"
ans=bracket_matching(exp)

if ans:
    
    print("expression has proper order of brackets")

else:
    
    print("brackets are misplaced")
    
```

# DECIMAL TO BINARY CONVERSION
for converting decimal to binary you divide number by 2 than write all remainder collectively in reverse order and this need of reverse order leads to use of stack

```python
def dec_bin(x):
    
    if x==0 or x==1:
        return x
    
    s=Stack()
    b=""
    
    while x>0:
        s.push(x%2)
        x=int(x/2)
        
    while not(s.is_empty()):
        b+=str(s.pop())
    
    return b
        
        
```

# Counter
a counter is same as dictionary except it is used to keep count and as such doesnot throw <mark> KeyError <mark>

```python
from collections import Counter
```

```python
cnt=Counter()
```

```python
for word in ['red','blue','green','red','yellow','blue','red','blue']:
    cnt[word]+=1   #if you were using dictionary it will throw keyerror try it atleast once
```

```python
cnt   #you can see whole dictionary
```

```python
cnt['red']    #retreive data as dictionary send key get value
```

```python
print(cnt.most_common())    # get all tuples of elements it has counted pass argument to function if you want top counts

print(cnt.most_common(2))   #lets say if you want only two most common wors pass 2
```
