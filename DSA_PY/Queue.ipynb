# Queue

```python
class Queue:
    def __init__(self):
        
        self.q=[]
    def enque(self,val):
        self.q=[val]+self.q
    def deque(self):
        return self.q.pop()
    def __str__(self):
        return str(self.q)
        
        

        
        
```

```python
q=Queue()
l=10
y=""
z=""
for i in range(l):
    q.enque(i)
    y+=str(i)+','
y=y.rstrip(',')

for i in range(l):
    z+=str(q.deque())+','
z=z.rstrip(',')    
print("enque order : ",y)
print("deque order : ",z)
```
