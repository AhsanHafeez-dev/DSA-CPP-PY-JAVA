```python
file_sizes=[]
for dir_path,dirs,files in os.walk("C:\\"):
    
    
    for f in files:
        file_full_path=os.path.join(dir_path,f)
        try:
            file_size=os.path.getsize(file_full_path)
        except Exception as e :
            continue
        file_sizes.append((file_full_path,file_size/(1024*1024)  ))
    
```

# Cards and Decks

```python
class card:
    def __init__(self,suit,val):
        self.suit=suit
        self.val=val
    def __str__(self):
        return str(self.val)+" of "+str(self.suit)
```

```python
class Deck:
    def __init__(self):
        self.deck=[]
        self.build()
    def build(self):
        
        for s in ["Spades","Hearts","Diamond","Clubs"]:
            for i in range(1,14):
                self.deck.append(card(s,i))
    def __str__(self):
        ret=""
        for c in self.deck:
            ret+=str(c)+"\n"
        return ret
        
        
```

```python
import random
```

```python
def shuffle(self):
    for i in range(0,len(self.deck)):
        s=random.rantint(0,i)
        self.deck[s],self.deck[i]=self.deck[i],self.deck[s]
Deck.shuffle=shuffle        
        
```

```python
def draw(self):
    c=self.deck[random.randint(0,len(self.deck) )  ]
    self.deck.remove(c)
    return c
Deck.draw=draw
```
