# Selection Sort
it finds minimum from rest of list (for first iteration it will consider whole list as rest) and swap it with current element giving us sorted list till i if you repeat it n(len of list) times it will sort the list

```python
def swap(lst,i,j):
    lst[i],lst[j]=lst[j],lst[i]
```

```python
def selection_sort(lst):
    """ takes the list as only parameter and sort it using selection sort algorithim """
    n=len(lst)
    
    #for each element in list (starting form left)
    
    for i in range(n):
        
        min_idx=i  # finding minimum ...
        
        #... in *rest* of list
        for j in range(i+1,n):
            
            #if current value is smaller than min than it our new minimum
            if  lst[j]<lst[min_idx]:
                min_idx=j
                
        #swap the minimum value with current value,we got sorted item till i!            
        swap(lst,i,min_idx)
        
```

```python
l=[x for x in reversed( range(999) )]
# print("before sorting : ",l)
%time selection_sort(l)
# print("after  sorting : ",l)
```

# Quick Sort

```python
import random
def quick_sort(lst,start,end):
    
    if end<=start:
        return
    
    less=start
    greater=end
    pivot=lst[random.randint(start,end)]
    while less<=greater:
        while lst[less]<pivot:
            less+=1
        while lst[greater]>pivot:
            greater-=1
        if less<=greater:
            swap(lst,less,greater)
            less,greater=less+1,greater-1
        
        
    quick_sort(lst,start,greater)
    quick_sort(lst,less,end)
    
```

# Real World Sorting
## Descending Sorts

```python
def selection_sort(lst,cmp):
    """ takes the list as only parameter and sort it using selection sort algorithim """
    n=len(lst)
    
    #for each element in list (starting form left)
    
    for i in range(n):
        
        desired_idx=i  # finding minimum ...
        
        #... in *rest* of list
        for j in range(i+1,n):
            
            #compare current value according to your need (for asc = lst[j]<lst[idx]) (for des = lst[j]>idx)
            if  cmp(lst[j],lst[desired_idx]):
                desired_idx=j
                
        #swap the desired value with current value,we got sorted item till i!            
        swap(lst,i,desired_idx)
        
```

```python
# predictae (discrete)
def compare_min(a,b):
    return a<b
def compare_max(a,b):
    return a>b
```

```python
l=[x for x in reversed( range(9) )]
print("before sorting : ",l)
selection_sort(l,compare_min)
print("after  sorting : ",l)
```

# Sorting objects of customs class

```python
class student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def __str__(self):
        return "("+self.name +","+str(self.age)+")"
```

```python
i=10
s1=student("subhan",9)
s2=student("ahsan",10)
s3=student("zohaib",8)
s4=student("anas",7)
l=[s1,s2,s3,s4]
```

```python
def age(x):
    return x.age
```

```python
def pr(x):
    
    s="["
    for i in x:
        s+=str(i)+','
    s=s.rstrip(',')
    print(s)
    
```

```python
print("before sorting : ",end="")
pr(l)
l.sort(key=age)
print("after  sorting : ",end="")
pr(l)
```
