# Doubly Linked List

### in linked list if you go even 1 step forward you can't go back you have to traverse whole list again
### what if you have million items in list so you have to traverse million items again
### its tidious and inefficient so whats the solution? !What if we have a way of going backward from any node
### solution is doubly linked list!!
### it is same as single or normal linked list except it has one extra reference variable previous or prev
### so now linked list is doubly connected you can go in both direction forward and backwad

```python
class Node:
    def __init__(self,val):
        
        self.next=None
        self.prev=None
        self.val=val
        
class DoublyLinkedList:
    
    def __init__(self):
        self.head=self.last=None
        self.len=0
        
        
    def __str__(self):
        
        s="["
        temp=self.head
        
        while temp is not None:
            print( hex(id(temp.prev)),hex(id(temp)),hex(id(temp.next)) )
            s+=str(temp.val)+','
            temp=temp.next
            
        s=s.rstrip(',')
        s+="]"
        
        return s
    
            
```

```python
def push(self,val):
    
    new_node=Node(val)
    self.len+=1
    
    if self.head is None:                       #insertion at head
        self.head=self.last=new_node
        return
    
    self.last.next=new_node
    new_node.prev=self.last
    self.last=new_node
    
DoublyLinkedList.push=push
```

```python
d=DoublyLinkedList()
l=5
# print(d)
for i in range(l):
    d.push(i)
# print(d)
```

```python
def pop(self):
    
    if self.head is None:                    #sanity check
        raise Exception("cannot pop from empty list")
        
    val=self.last.val    
    self.len-=1
    
    if self.head.next is None:               #case 1: one node only
        self.head=self.last=None             # automtatic garbage collection  
    
    else:                                   # case 2 : more than one node
        self.last.prev.next=None
        self.last=self.last.prev           # automtatic garbage collection
    
    return val

DoublyLinkedList.pop=pop
```

```python
def remove(self,val):
    
    if self.head is None:                    #sanity check
        raise Exception("cannot remove from empty list")
        
    if self.head.val==val:                # removing head
        self.head.next.prev=None
        self.head=self.head.next
        
    else:                                # removing anynode other than head
        
        current=self.head
        
        while current is not None  and current.val!=val :
            current=current.next
            
        if current is None:
            print("Value not found")
            return
            
        current.prev.next=current.next
        
        if current.next is not None:                              #checking for last node deletion case 
            current.next.prev=current.prev
            
        else:
            self.last=current.prev
    self.len-=1
            
DoublyLinkedList.remove=remove            
```

```python
def insert(self,idx,val):
    
    new_node=Node(val)
    
    if self.head is None:
        self.head=self.last=new_node
    if idx==0:
        new_node.next=self.head
        self.head.prev=new_node
        self.head=new_node
        self.len+=1
        
    elif idx>=self.len:
        self.last.next=new_node
        new_node.prev=self.last
        self.last=new_node
        
    else:
        current=self.head
        i=0
        
        while i<idx:
            current=current.next
            
        current.next.prev=new_node
        current.next=new_node
        
        new_node.next=current.next
        new_node.prev=current
    self.len+=1
DoublyLinkedList.insert=insert        
```

```python
def get(self,idx):
    if idx>=self.len or idx<0 or self.head is None:
        raise IndexError("list has total ",self.len-1," indices you are trying to access index ",idx)
    current=self.head 
    i=0
    while i<idx:
        current=current.next
    return current.val
DoublyLinkedList.get=get
```
