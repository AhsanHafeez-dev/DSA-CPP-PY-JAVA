# Linked List
Linked list is collection of nodes containing the data and pointer(reference variable) to the next item in list.
The linked list itself is just a pointer to head using which you can traverse through list
## Data Structure

```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None
class LinkedList():
    def __init__(self):
        self.head=self.last=None
        self.len=0
```

## Operations
## 1.Push
Three are two cases in push:
    1.When list is empty (0 node self.head is empty)
    2.When list is not empty (atleast one node self.head is not None)

```python
def push(self,val):
    
    new_node=Node(val)
    self.len+=1
    
    if self.head is None:
        self.head=self.last=new_node
        return
    
    #instead of using last pointer you can also traverse through whole list for each push(space time tradeoff)
    self.last.next=new_node
    self.last=new_node
    
    
LinkedList.push=push
```

```python
def __str__(self):
    
    s="["
    t=self.head
    
    while t is not None:
        s+= str(t.val)+','
        t=t.next
    
    s=s.rstrip(',')
    s+="]"
    
    
    return s


LinkedList.__str__=__str__
```

```python
l=LinkedList()

for i in range(5):
    l.push(i)
    
print(l)
```

## 2.Pop
pop is deleting last element of data structure(here list)
pop can have 2 cases as well:
    1.when you have only one node (you have to pop head)
    2.when there are more than one node 

```python
def pop(self):
    
    if self.head is None: #sanity check
        raise Exception("Cannot pop from empty list")
        
    self.len-=1 
    
    if self.head.next is None:  #case 1 if there is only one node
        val=self.head.val
        self.head=self.last=None
        return val

    #case 2 any node other than head
    
    t=self.head
    val=self.last.val  #retreiving last value for returning
    
    while t.next!=self.last:
        t=t.next
        
    self.last=t  #setting new last
    t.next=None  #make sure no one points to it so it can be garbage collected
    
    return val


LinkedList.pop=pop        
```

```python
print("before popping : ",l)
l.pop()
print("after popping  : ",l)
```

## Insertion
Insertion is placing the new value or node in between structure<br>
Insertion can have 3 case:<br>
    1.when list is empty<br>
    2.when inserting at 0 index<br>
    3.inserting at last<br>
    4.inserting in between list<br>
first two cases are just the same as push thats why consider them as one
    

```python
def insert(self,index,val):
    
    new_node=Node(val)
    
    
    if self.head is None:  #sanity check
        self.head=self.last=new_node
        self.len+=1
        return
    
    if index ==0:  #insertion at head is different
        new_node.next=self.head
        self.head=new_node
        self.len+=1
        return
    
    if index>=self.len:   #if user giving index greater than legth we will insert at last
        self.last.next=new_node
        self.last=new_node
        self.len+=1
        return
    print(index,self.len)
    current=self.head
    i=0
    while i<index:
        current=current.next
        
    new_node.next=current.next #first connect new node to next node to current node
    current.next=new_node
    self.len+=1
    
    
LinkedList.insert=insert
```

```python
def get(self,idx):
    if self.head is None :  #check if list is empty sanity check
        raise IndexError("List is empty")
    
    if idx>=self.len or idx < 0:  #check if user asked for valid index
        raise IndexError("Index Out of Bound list has ",self.len-1,"indices you are trying to access ",idx)
    t=self.head
    i=0
    while i!=idx:
        t=t.next
        i+=1
    return t.val
        
        
```

```python
def remove(self,val):
    self.len-=1
    if self.head is None:            #sanity check
        raise Exception("List is Empty")
        
    if self.head.val==val:          #case 1 if we have to remove head we should reset self.head
        self.head=self.head.next
        
    
        
    current=self.head
    while True:
        prev=current
        current=current.next
        if(current.val==val) or current is None:
            break
        
    if current is None:          #if last node is removed self.last should be set
        self.len+=1
        print("value not found")
        
    prev.next=current.next
    if current.next is None:
        self.last=prev
        
    
    
LinkedList.remove=remove    
```
