```python
def distance(a,b):
    return sqrt(a**2+b**2)
def estimate_pi(rep):
    darts=rep
    din=0
    for i in  range(rep):
        x,y=r.random(),r.random()
        if distance(x,y)<1:
            din+=1
            
    pi=(din/rep)*4
    return pi
```

## 2.Random  Walks

```python
def get_action():
    return r.choice( [ (0,1), (0,-1), (1,0), (-1,0), ]  )
def random_walks(n):
    """return path after 'n' block random walks """
    path=[]
    x,y=0,0
    
    for i in range(n):
        (dx,dy)=get_action()
        x+=dx
        y+=dy
        path.append((x,y))
        
    return path
        
```

```python
n=200
walk=random_walks(n)
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None,figsize=(15,10))
point_color=list(range(n))

x_number =[x[0] for x in walk]
y_number=[x[1] for x in walk]
plt.scatter(x_number,y_number,marker=".",s=40,c=point_color)
plt.show()
```
