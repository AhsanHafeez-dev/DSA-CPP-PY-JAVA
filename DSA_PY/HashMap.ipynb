# HashMap
A HashMap is a specific implementation of a Map commonly used in various programming languages. It provides an efficient way to store and retrieve data through key-value pairs. Key features of a HashMap include:

### Constant Time Complexity: 
O(1) for most operations like get, put, and remove.
### Capacity and Load Factor:
Initial size and the factor at which the map's capacity increases when the current capacity exceeds a certain threshold.

# Hashing
Hashing is the process of converting an input (or 'key') into a fixed-size string of bytes, typically a hash code. The hash function is the method used to perform this conversion. The main properties of a good hash function are:

### Deterministic: 
The same input always produces the same output.
### Efficient:
Computes the hash code quickly.
### Uniform Distribution:
Distributes keys uniformly across the array to minimize collisions.
### Low Collision Probability: 
Produces a unique hash code for each unique key, as much as possible.

## Pigeonhole Principle
The pigeonhole principle is a simple yet powerful concept in combinatorics and mathematics. It states:

"If n items are put into ð‘š  containers, with n>m, then at least one container must contain more than one item."

In other words, if you have more items than containers and you distribute the items among the containers, at least one container will hold more than one item. This principle is often used in proofs and arguments to demonstrate that collisions or overlaps are inevitable in certain scenarios.

```python
def insert(self,key,val):
    k=self.Hash(key)
    if self.map[k]==None:
        self.map[k]=[[key,val]]
    else:
        for i in self.map[k]:
            if i[0]==key:
                i[1]=val
                return
        self.map[k].append([key,val])
HashMap.insert=insert        
```

```python
def delete(self,key):
    key_hash=self.Hash(key)
    if self.map[key_hash] is not None:
        for i in range(len(self.map[key_hash])):
            
            if self.map[key_hash][i][0]==key:
                self.map[key_hash].pop(i)
                return
    raise KeyError(str(key))
HashMap.delete=delete
```
