## Factorial
### mathematically
    factorial is product of number with factorial of number -1

```python
def factorial(n):
    if n<=1:      #try commenting this base case and see what happens
        return 1
    return n * factorial(n-1)
```

```python
for i in range(1,7):
    print("factorial of ",i,"is : ",factorial(i))
```

## Fibonacci Series
we are not going to discuss whats its practical usage but just search golden ration its basically also used to see where will user focus when he sees anything

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
```

```python
n=20
%time fib(n)
```

```python
def fib(n):
    a=0
    b=1
    for i in range(n):
        a,b=b,a+b
    return a
```

```python

%time fib(n)
```

```python
def fib2(n,a=0,b=1):
    if n==0:
        return a
    return fib2(n-1,b,a+b)     # tail recursion
```

```python
%time fib2(n)
```

## Tower Of Hanoi

```python
def print_move(src,dest,aux,n):
    print("moving disk ",n," from",src," to ",dest," using ",aux," as auxilary")
```

```python
def tower_of_hanoi(src,dest,aux,n):
    if n==1:
        print_move(src,dest,aux,n)
        return
    tower_of_hanoi(src,aux,dest,n-1)
    print_move(src,dest,aux,n)
    tower_of_hanoi(aux,dest,src,n-1)
    
```

```python
tower_of_hanoi('a','c','b',3)
```

## sum over list
sum of list is simply sum of head value + sum of rest of list

```python
def sum_lst(lst):
    
    if len(lst)==0:
        return 0
    
    else :
        return lst[0]+sum_lst(lst[1:])   #what it is head or tail recursion?
```

```python
lst=[1,2,3,4,5]
sum_lst(lst)
```
