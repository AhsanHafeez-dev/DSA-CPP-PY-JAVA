## Buble sort
start by moving maximum element than second highest and so on but it is not used anywhere bcz of inefficiency and n^2 time complexity but it is a good starting point as it is brute force approach

```python
def swap(lst,i,j):
    lst[i],lst[j]=lst[j],lst[i]
```

```python
def buble_sort(lst):
    
    n=len(lst)
    # print(n)
    
    # outerloop goes over whole thing 'n' times
    # because each time one ' highest ' will move to end so we need n pases to sort whole thing
    for i in range(n-1):
        
        # try to buble one thing(hightest) up 
        for j in range( 0, (n-i)-1):
            
            # compare pairs ,move higher one up (highest will reach end)
            if lst[j]>lst[j+1]:
                swap(lst,j+1,j)
                
```

```python
l=[x for x in reversed(range(10))]
```

```python
print(l)
buble_sort(l)
print(l)
```
