```python
class Node:
    def __init__(self,val):
        
        self.next=None
        self.prev=None
        self.val=val
        
class CircularLinkedList:
    def __init__(self):
        self.head=self.last=None
        self.len=0
    def __str__(self):
        
        
        if self.head is None:
            return "[]"
        
        s="["
        
        
        
        s+=str(self.head.val)+","

        current=self.head.next

        while current!=self.head:
            s+=str(current.val)+','
            current=current.next

        s=s.rstrip(',')
        s+="]"

        return s
```

```python
def push(self,val):
    
    new_node=Node(val)
    self.len+=1
    
    if self.head is None:                       #insertion at head
        self.head=self.last=new_node
        self.last.next=self.head.prev=new_node
        
        
        return
    
    self.last.next=new_node
    new_node.prev=self.last
    self.last=new_node
    new_node.next=self.head
    self.head.prev=new_node
    
CircularLinkedList.push=push
```

```python
d=CircularLinkedList()
l=5
print(d)
for i in range(l):
    d.push(i)
print(d)
```

```python
def pop(self):
    
    if self.head is None:                    #sanity check
        raise Exception("cannot pop from empty list")
        
    val=self.last.val    
    self.len-=1
    
    if self.head.next is None:               #case 1: one node only
        self.head=self.last=None             # automtatic garbage collection  
    
    else:                                   # case 2 : more than one node
        self.last.prev.next=self.head
        self.last=self.last.prev           # automtatic garbage collection
        self.head.prev=self.last
    
    return val

CircularLinkedList.pop=pop
```

```python
def remove(self,val):
    
    if self.head is None:                    #sanity check
        raise Exception("cannot remove from empty list")
    
    if self.head.val==val:                # removing head
        self.head.next.prev=None
        self.head=self.head.next
        self.last.next=self.head
        
    else:                                # removing anynode other than head
        
        current=self.head.next
        
        while current!=self.head   and current.val!=val :
            
            current=current.next
         
        
        if current==self.head:
            print("value not found")
            return
        
        current.prev.next=current.next
        
        if current.next == self.head:       #checking for last node deletion case  to update self.last
            
            
            self.last=current.prev
            
            
        current.next.prev=current.prev
            
    self.len-=1
            
CircularLinkedList.remove=remove            
```

```python
def setHeadLast(self):
    self.head.prev=self.last
    self.last.next=self.head
CircularLinkedList.setHeadLast=setHeadLast
```

```python
def insert(self,idx,val):
    
    new_node=Node(val)
    
    if self.head is None:
        self.head=self.last=new_node
        self.last.next=self.head.prev=new_node
        
    if idx==0:
        new_node.next=self.head
        self.head=new_node
        self.setHeadLast()
        
        self.len+=1
        
    elif idx>=self.len:
        print("last insertion")
        self.last.next=new_node
        new_node.prev=self.last
        self.last=new_node
        self.setHeadLast()
        
    else:
        current=self.head
        i=0
        
        while i<idx:
            current=current.next
            
        current.next.prev=new_node
        current.next=new_node
        
        new_node.next=current.next
        new_node.prev=current
    self.len+=1
CircularLinkedList.insert=insert        
```
