```python
class Node:
    def __init__(self,val):
        self.val=val
        self.next=None
    
class LinkedList:
    def __init__(self):
        self.head=None
        self.last=None
        self.len=0
    def insert(self,idx,val):
        
        
        if idx<0:
            print("Index Error")
            return 
        new_node=Node(val)
        
        if self.head is None:
            print("case 1 : adding first node in list")
            self.head=self.last=new_node
            self.len+=1
            return
        if idx==0:
            new_node.next=self.head
            self.head=new_node
            print("case 1 : insertion at head")
            return
        
        if idx==self.len:
            print("case 2 : adding at last element")
            self.last.next=new_node
            self.last=new_node
            self.len+=1
            return
        print("case 2 : adding at mid")
        temp=self.head
        i=0
        while i<idx and temp.next is not None:
            temp=temp.next
            i+=1
        if temp.next:
            new_node.next=temp.next.next
        
        temp.next=new_node
        self.len+=1
        
    
    def __str__(self):
        
        ret="["
        temp=self.head
        while temp is not None:
            ret+=str(temp.val)+','
            
            temp=temp.next
        ret=ret.rstrip(',')
        ret+="]"
        return ret
    def add(self,val):
        self.insert(self.len,val)
    def delete(self,idx):
        if idx>=self.len or self.len==0 or idx<0: 
            print("index error")
            return
        if self.head.next is None:
            print("case 1 : only one node in list")
            t=self.head.val
            self.head=None
            return t
        if idx==0:
            print("removing head")
            t=self.head.next.val
            self.head=self.head.next
            return t
        print("case 2 : more than one nodes")
        temp=self.head
        i=0
        while i<idx-1:
            i+=1
            temp=temp.next
        
        ret=temp.next.val
        temp.next=temp.next.next
        return ret
    def pop(self):
        
        self.delete(self.len-1)
        
```

```python
class Stack:
    def __init__(self):
        self.stack=[]
    def push(self,val):
        self.stack.append(val)
    def pop(self):
        if len(self.stack)>0:
            return self.stack.pop()
        else :
            raise IndexError("StackUnderflow!!stack is empty")
    def peek(self):
        if len(self.stack)>0:
            return self.stack[-1]
        else :
            raise IndexError("stack is empty")
    def is_Empty(self):
        return len(self.stack)==0
```

```python
def bracket(expression):
    opening = '({['
    closing = ')}]'
    matches=dict(zip(opening,closing))
    
    stack=Stack()
    for e in expression:
        if e in matches:
            stack.push(matches[e])
        elif e in matches.values() and not stack.is_Empty() and e!=stack.pop():
            print("expression mis match")
            return False
    return stack.is_Empty()
```

```python
def dec(bi):
    if bi==0:
        return bi
    stack=Stack()
    while bi>0:
        stack.push(bi%2)
        bi=bi//2
        
    s=''
    while not stack.is_Empty():
        s+=str(stack.pop())
    return int(s)
```

```python
def find_min(lst):
    if len(lst)>0:
        idx=0
        
        for i in range(1,len(lst)):
            
            if lst[i] < lst[idx]:
                idx=i
        return idx
    return None
```

```python
class TreeNode:
    def __init__(self,val):
        self.val=val
        self.right=self.left=None
    def traverse(self,fn):
        fn(self.val)
        if self.left:
            self.left.traverse(fn)
        if self.right:
            self.right.traverse(fn)
        
    
```

```python
def bfs(root,fn):
    if root is not None:
        to_process=[root]
        while to_process:
            node=to_process.pop(0)
            if node.right:
                to_process.append(node.right)
            if node.left:
                to_process.append(node.left)
            fn(node.val)
```
